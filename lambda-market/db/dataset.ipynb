```python
combo = all_vendors.groupby("id").agg({"price": "max", "stockLevel": "sum"}).merge(df[["classId", "productType"]], right_on="classId", left_index=True, how="right")
```

```python
multiplier = {
    "Fruit": 2,
    "Packages": 4,
    "Vegetables": 2,
    "Reptiles": 100,
    "Oddity": 1000,
}
```
